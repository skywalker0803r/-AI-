預測目標:DeSOx_1st 其餘為特徵

1. 先將數據select_df依據時間排序。

2. 自第1001筆開始，取用前1000筆數據建立rf模型，評估建模數據的"預測誤差絕對值"，計算前述數值的第80百分位數當成"門檻值"。

3. 運用2.的模型執行第1001筆的預測。

4. 比對3.的結果之"預測誤差絕對值"。若此數值超過"門檻值"，則第1002筆預測重複2.＆3.步驟；若此數值不超過"門檻值"，直接運用2.的模型執行第1002筆的預測。

5. 之後以此類推....

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from tqdm import tqdm
import warnings
warnings.filterwarnings("ignore")

# ======================== 使用者自定義超參數 ==========================
PERCENTILE_THRESHOLD = 95     # 門檻百分位 (預設: 80)
PREDICTION_INTERVAL = 300     # 每次預測跳躍間隔 (預設: 100，可設為 300、500 等)
MAX_STEPS = 1000              # 最多預測幾步
WINDOW_SIZE = 1000            # 每次用多少資料訓練
# =====================================================================

# 1. 資料排序
select_df = select_df.sort_index().reset_index(drop=True)

# 2. 設定欄位
target_col = y_col
feature_cols = [col for col in select_df.columns if col not in ['timestamp', target_col]]
start_idx = WINDOW_SIZE
total_steps = len(select_df) - start_idx

# 3. 儲存預測結果
predictions = []
abs_errors = []
thresholds = []
indices = []

# 4. 初始化進度條
pbar = tqdm(total=MAX_STEPS)

# 5. 動態預測迴圈
i = start_idx
while i < len(select_df):
    if pbar.n >= MAX_STEPS:
        print(f"已達 {MAX_STEPS} 步，提前結束迴圈")
        break

    # 訓練資料
    train_df = select_df.iloc[i - WINDOW_SIZE:i]
    train_X = train_df[feature_cols]
    train_y = train_df[target_col]

    # 訓練模型
    model = RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1)
    model.fit(train_X, train_y)

    # 計算門檻
    train_preds = model.predict(train_X)
    train_errors = np.abs(train_preds - train_y)
    threshold = np.percentile(train_errors, PERCENTILE_THRESHOLD)

    # 預測當前資料
    test_row = select_df.iloc[i]
    test_X = test_row[feature_cols].values.reshape(1, -1)
    true_y = test_row[target_col]
    pred_y = model.predict(test_X)[0]
    error = abs(pred_y - true_y)

    # 紀錄
    predictions.append(pred_y)
    abs_errors.append(error)
    thresholds.append(threshold)
    indices.append(i)

    if error > threshold:
        # 誤差太大 → 下一筆繼續預測
        i += PREDICTION_INTERVAL
        pbar.update(1)
    else:
        # 誤差小 → 預測下一筆（額外 bonus 預測）
        next_i = i + PREDICTION_INTERVAL
        if next_i < len(select_df):
            next_row = select_df.iloc[next_i]
            next_X = next_row[feature_cols].values.reshape(1, -1)
            next_true_y = next_row[target_col]
            next_pred_y = model.predict(next_X)[0]
            next_error = abs(next_pred_y - next_true_y)

            predictions.append(next_pred_y)
            abs_errors.append(next_error)
            thresholds.append(threshold)
            indices.append(next_i)

            pbar.update(2)
            i += 2 * PREDICTION_INTERVAL
        else:
            pbar.update(1)
            i += PREDICTION_INTERVAL

pbar.close()

# 6. 輸出結果表
result_df = select_df.loc[indices].copy()
result_df['prediction'] = predictions
result_df['abs_error'] = abs_errors
result_df['threshold'] = thresholds

result_df.head()
```
